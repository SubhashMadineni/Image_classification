## Data Preparation

```python
X_train = X_train.astype('float32')
X_test = X_test.astype('float32')
```

```python
num_cat = 10
```

```python
y_train
```

```python
import keras
y_train= keras.utils.to_categorical(y_train, num_cat)
```

```python
y_train
```

```python
y_test= keras.utils.to_categorical(y_test, num_cat)
```

```python
y_test
```

```python
X_train = X_train/255
X_test = X_test/255
```

```python
X_train
```

```python
X_train.shape
```

```python
Input_shape= X_train.shape[1:]
```

```python
Input_shape
```

## Training the model

```python
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D,AveragePooling2D, Dense, Flatten, Dropout
from keras.optimizers import adam
from keras.callbacks import TensorBoard
```

```python
cnn_model = Sequential()
cnn_model.add(Conv2D(filters = 64, kernel_size = (3,3), activation = 'relu',input_shape = Input_shape))
cnn_model.add(Conv2D(filters = 64, kernel_size = (3,3), activation = 'relu'))
cnn_model.add(MaxPooling2D(2,2))
cnn_model.add(Dropout(0.3))

cnn_model.add(Conv2D(filters = 128, kernel_size = (3,3), activation = 'relu'))
cnn_model.add(Conv2D(filters = 128, kernel_size = (3,3), activation = 'relu'))
cnn_model.add(AveragePooling2D(2,2))
cnn_model.add(Dropout(0.3))

cnn_model.add(Flatten())

cnn_model.add(Dense(units = 512, activation='relu'))

cnn_model.add(Dense(units = 1024, activation='relu'))

cnn_model.add(Dense(units = 10, activation = 'softmax'))
```

```python
cnn_model.compile(loss = 'categorical_crossentropy', optimizer = keras.optimizers.rmsprop(lr= 0.001), metrics = ['accuracy'])
```

```python
history = cnn_model.fit(X_train,y_train, batch_size= 32, epochs = 4, shuffle = True)
```

## Evaluating the model

```python
evaluation = cnn_model.evaluate(X_test,y_test)
print('Test Accuracy  : {}'.format(evaluation[1])  )
```

```python
predicted_classs = cnn_model.predict_classes(X_test)
predicted_classs
```

```python
y_test = y_test.argmax(1)
y_test
```

```python
L = 7
W = 7
fig,axes = plt.subplots(L, W, figsize = (12,12))
axes = axes.ravel()
for i in np.arange(0,L*W):
    axes[i].imshow(X_test[i])
    axes[i].set_title('prediction = {}\n True = {}'.format(predicted_classs[i],y_test[i]))
    axes[i].axis('off')
plt.subplots_adjust(wspace= 1)
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns


cm = confusion_matrix(y_test, predicted_classs)
cm
plt.figure(figsize = (10,10))
sns.heatmap(cm, annot = True)
```

## Save the model

```python
import os
directory = os.path.join(os.getcwd(), 'Saved_Models')

if not os.path.isdir(directory):
    os.makedirs(directory)
model_path = os.path.join(directory, 'Keras_cifar10_trained_modek.h5')
cnn_model.save(model_path)
```

## image augumentation

```python
import keras
from keras.datasets import cifar10
(X_train,y_train), (X_test, y_test) = cifar10.load_data()
```

```python
X_train = X_train.astype('float32')
X_test = X_test.astype('float32')
```

```python
X_train.shape
```

```python
n = 8
X_train_sample = X_train[:n]
```

```python
X_train_sample.shape
```

```python
from keras.preprocessing.image import ImageDataGenerator
dataget_train = ImageDataGenerator(rotation_range= 90)
dataget_train.fit(X_train_sample)
```

```python
from scipy.misc import toimage

fig = plt.figure(figsize = (20,2))
for x_batch in dataget_train.flow(X_train_sample, batch_size = n):
    for i  in range(0,n):
        ax= fig.add_subplot(1, n, i+1)
        ax.imshow(toimage(x_batch[i]))
    fig.suptitle('Augmented images (rotated 90 degrees)')
    plt.show()
    break;
```
